```python
target_col = "Heart Disease"

y = train[target_col].map({
    "Absence": 0,
    "Presence": 1
})

X = train.drop(columns=[target_col])
```

```python
cat_cols = [c for c in X.columns if X[c].nunique() < 10]
num_cols = [c for c in X.columns if c not in cat_cols]
```

```python
print("Heart Disease" in cat_cols)
```

```python
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)
test_encoded = pd.get_dummies(test, columns=cat_cols, drop_first=True)
```

```python
print(cat_cols)
print(set(cat_cols) - set(test.columns))
```

```python
print(X_encoded.shape)
print(test_encoded.shape)
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss
import numpy as np

oof = np.zeros(len(X_encoded))
test_preds = np.zeros(len(test_encoded))

for fold, (tr_idx, val_idx) in enumerate(skf.split(X_encoded, y)):
    X_tr, X_val = X_encoded.iloc[tr_idx], X_encoded.iloc[val_idx]
    y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

    model = LogisticRegression(max_iter=2000)
    model.fit(X_tr, y_tr)

    val_pred = model.predict_proba(X_val)[:, 1]
    test_pred = model.predict_proba(test_encoded)[:, 1]

    oof[val_idx] = val_pred
    test_preds += test_pred / skf.n_splits

    print(f"Fold {fold} LogLoss:", log_loss(y_val, val_pred))

print("\nOverall CV LogLoss:", log_loss(y, oof))
```

```python
# If not installed:
# !pip install xgboost

from xgboost import XGBClassifier
from sklearn.metrics import log_loss
import numpy as np

oof = np.zeros(len(X_encoded))
test_preds = np.zeros(len(test_encoded))

for fold, (tr_idx, val_idx) in enumerate(skf.split(X_encoded, y)):
    print(f"\nTraining Fold {fold}")

    X_tr, X_val = X_encoded.iloc[tr_idx], X_encoded.iloc[val_idx]
    y_tr, y_val = y.iloc[tr_idx], y.iloc[val_idx]

    model = XGBClassifier(
        n_estimators=1500,
        learning_rate=0.01,
        max_depth=4,
        min_child_weight=1,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_lambda=1,
        objective="binary:logistic",
        eval_metric="logloss",
        random_state=42,
        tree_method="hist"   # faster
    )

    model.fit(
        X_tr, y_tr,
        eval_set=[(X_val, y_val)],
        verbose=False
    )

    val_pred = model.predict_proba(X_val)[:, 1]
    test_pred = model.predict_proba(test_encoded)[:, 1]

    oof[val_idx] = val_pred
    test_preds += test_pred / skf.n_splits

    print("Fold LogLoss:", log_loss(y_val, val_pred))

print("\nOverall CV LogLoss:", log_loss(y, oof))
```
